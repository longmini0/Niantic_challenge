```python
# fill the non type 2 and combine type 1 and type 2
poke_stats['Type_2'] = poke_stats.apply(lambda x: x['Type_1'] if str(x['Type_2']) == 'nan' else x['Type_2'], axis = 1)
poke_stats['Combined_Type'] = poke_stats['Type_1'] + ' ' + poke_stats['Type_2']
# change the column order
poke_stats = poke_stats[['Name', 'Type_1', 'Type_2', 'Combined_Type', 'Total', 'HP', 'Attack',
                      'Defense', 'Sp_Atk', 'Sp_Def', 'Speed']]
```

Looks like there is a total 143 unique combined type of pokemons. I would like to calculate the mean value of each stat.

```python
combined_stats = poke_stats.groupby('Combined_Type').mean()
```

```python
combined_stats
```

Let's plot top-15 mean value for each stat from combined type

```python
types = ['HP', 'Attack', 'Defense', 'Sp_Atk', 'Sp_Def', 'Speed']
i = 1
c = 0
palettes = sns.color_palette()[1:7]

plt.figure(figsize = (24,24))
for ty in types:
    plt.subplot(3,2,i)
    i = i+1
    sns.barplot(x = ty, y = 'Combined_Type',
               data = combined_stats.reset_index().sort_values(ty, ascending = False).iloc[:15],
               color = palettes[c])
    c = c+1
    
    plt.title(str('Top 15 of mean of ' + ty))
```

We can see several type-combinations such as [Ghost-Dragon, Dragon-Fire, Steel-Dragon, Psychic-Ghost] are in the top 15 of some base stats

We can see 'Steel-Dragon' has the top total mean value among all 143 combined type.

## Q3. Model options for legendary prediction

Since we want to predict whether the Pokemon is a legendary or not, this problem is 2-class classfication problem, and the dataset is not too large. Therefore, I would like to use 3 machine learning models below.

### 1) Logistic Regression
##### Pros
It is probably the most widely used machine learning algorithm. The output from logistic regression has a nice probablilistic interpretation, and the algorithm can be regularized to avoid overfitting using L1(Lasso) or L2(Ridge) regularization.
##### Cons
It tends to underperform when there are multiple or non-linear boundaries. It is not flexible enough to naturally capture more complicated relationships.

### 2) Decision Tree
##### Pros 
Decision Tree the algorithm that is easy to interpret and explain. It learns hierarchically by repeatedly splitting the dataset into separate branches that maximize the information gain of each split. This structure allows decision trees to learn non-linear relationships. Robust to outliers, scalable, and able to naturally model non-linear decision boundaries.
##### Cons
It can easily overfit. Since individual trees keep brancjhing until memorizing the training data, they are prone to overfitting.

### 3) Random Forests

##### Pros
Randon Forest is machine learning algorithm that is easy and flexible to use in order to achieve great results. It trains each tree independently using random sample of the data, so the trained model is more robust than a single decision tree and less likely to overfit. Also, we can get lower classfication error and better f-scores than decision trees. It is also great model for the dataset where it is uneven and has missing variables.

##### Cons
It also has overfitting risk. If the data has categorical variables with different levels of attributes, random forest algorithm can pose a prediction risk.

```python
X = pokemon.drop(columns='isLegendary')
y = pokemon['isLegendary']
```

Let's split first

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=3) 
```

Let's split the data into categorical and numerical

```python
col_cat = ['Pr_Male', 'Body_Style', 'hasGender', 'hasMegaEvolution']
col_num = ['Total', 'Height_m', 'number_of_types']

X_train_cat = X_train[col_cat]
X_train_num = X_train[col_num]
X_test_cat = X_test[col_cat]
X_test_num = X_test[col_num]
```

```python
from sklearn.preprocessing import OneHotEncoder # distribute 0 and 1
from sklearn.pipeline import make_pipeline
```

```python
scaler_cat = make_pipeline(OneHotEncoder())
X_train_cat_enc = scaler_cat.fit_transform(X_train_cat)
X_test_cat_enc = scaler_cat.transform(X_test_cat) 

scaler_num = make_pipeline(StandardScaler())
X_train_num_scaled = scaler_num.fit_transform(X_train_num)
X_test_num_scaled = scaler_num.transform(X_test_num)
```

We should apply these transformations on the training and testing sets

```python
import numpy as np
from scipy import sparse

X_train_scaled = sparse.hstack((X_train_cat_enc,
                                sparse.csr_matrix(X_train_num_scaled))) # make the same data structure 
X_test_scaled = sparse.hstack((X_test_cat_enc,
                               sparse.csr_matrix(X_test_num_scaled)))
```

In order to avoid overfitting, I oversampled the minority data points using SMOTE module in the training dataset

## Machine Learning

The machine learning model I chose is Random Forest. As I said above, it is an ensemble classifier which is constructed by many decision trees. Random Forest generally prevents overfitting and lowers variace thanks to its random nature. 

### Training the random forest

'n_estimators' argument determines the number of decision tress used in the forest.

'max_depth' of each tree in the forest is set to 3 levels

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_jobs=-1, n_estimators=300, random_state=2, max_depth=3)
rf.fit(X_train_scaled, y_train)
```

### Evaluation

```python
predicted_labels = rf.predict(X_test_scaled)
print("Train Score : ", rf.score(X_train_scaled, y_train))
print("Test Score : ", rf.score(X_test_scaled, y_test))
print("Accuracy : ", accuracy_score(y_test, predicted_labels))
```

```python
# Confusion matrix
from sklearn.metrics import confusion_matrix
cnf_matrix = confusion_matrix(y_test, predicted_labels)

p = sns.heatmap(pd.DataFrame(cnf_matrix), annot = True, fmt = 'd')
plt.xlabel("Predicted Values")
plt.ylabel("Actual Values")
plt.title("Confusion Matrix for Randon Forest Classifier")
```

```python
# roc auc curve for random forest classifier
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
pred_prob = rf.predict_proba(X_test_scaled)[:,1]
fpr, tpr, thresholds = roc_curve(y_test, pred_prob)
print("ROC-AUC Score : ", roc_auc_score(y_test, pred_prob))
plt.plot(fpr,tpr)
plt.plot([0,1],[0,1])
plt.xlabel("False positive rate")
plt.ylabel("True positive rate")
```
